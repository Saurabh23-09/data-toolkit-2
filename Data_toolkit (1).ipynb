2-  How does broadcasting work in NumPy?
 <!-- Broadcasting is a powerful mechanism in NumPy that allows you to perform arithmetic operations between arrays of different shapes, as long as they are compatible. It eliminates the need to write explicit loops, making your code more concise and efficient.

How it works:

Shape Compatibility:
      Broadcasting works by comparing the shapes of the arrays involved in an operation, starting from the trailing dimensions (rightmost) and moving backward.

Rules for Compatibility:
           Two dimensions are compatible if:

  They are equal in size.
   One of them is 1.
Automatic Expansion:
           If the dimensions are not equal, NumPy automatically "stretches" or "broadcasts" the smaller array to match the shape of the larger array. This expansion is virtual and doesn't involve actual data duplication, making it memory-efficient.

Element-
      wise Operation: Once the shapes are compatible, the operation is performed element-wise between the corresponding elements of the arrays. -->

21-  Why is NumPy’s array slicing faster than Python’s list slicing?

<!-- 1. Data Storage:

NumPy arrays:
     Store data in contiguous memory locations, meaning elements of the same array are stored next to each other in memory. This allows for efficient access and manipulation of data, as the location of each element can be easily calculated.

Python lists:

2. Data Type:

NumPy arrays:
    Are homogeneous, meaning they contain elements of the same data type. This allows for optimized operations and vectorization, where operations can be applied to entire arrays at once, leveraging low-level, compiled code (often written in C or Fortran) for speed.

Python lists:   
   Can contain elements of different data types. This flexibility requires Python to perform type checking and conversions during operations, which can add overhead and slow down slicing.

3. Optimized Operations:

NumPy:
   Is built for numerical computations and provides optimized functions for array operations, including slicing. These functions are often implemented in low-level languages (C or Fortran) for speed.

Python:
    List slicing involves creating a new list with copies of the selected elements. This process of copying elements can be slow, especially for large lists.
4. Vectorization:

NumPy:
    Supports vectorization, which allows operations to be applied to entire arrays at once without explicit looping in Python. This significantly speeds up computations, including slicing.

Python:
   List slicing often involves implicit loops to iterate over elements and create a new list, which can be slower than NumPy's vectorized operations. -->

1- How do you create a 2D NumPy array and calculate the sum of each row?

```python
# Importing the NumPy library with an alias 'np'
import numpy as np

# Creating a NumPy array 'x' with a 2x2 shape
x = np.array([[0, 1], [2, 3]])

# Printing a message indicating the original array 'x'
print("Original array:")
print(x)

# Calculating and printing the sum of all elements in the array 'x' using np.sum()
print("Sum of all elements:")
print(np.sum(x))

# Calculating and printing the sum of each column in the array 'x' using np.sum() with axis=0
print("Sum of each column:")
print(np.sum(x, axis=0))

# Calculating and printing the sum of each row in the array 'x' using np.sum() with axis=1
print("Sum of each row:")
print(np.sum(x, axis=1))
```
